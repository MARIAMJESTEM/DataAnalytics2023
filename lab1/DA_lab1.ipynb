#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import halfnorm, norm
from scipy import stats
```

#### Ex1. Generate arrays of:

```python
#1000 samples of uniform distribution on interval [0,1]
ex1_1 = np.random.uniform(0,1,1000)

#1000 samples of normal distribution with mean =0 and standard deviation =1.
ex1_2 = np.random.standard_normal(1000)

#1000 samples of normal distribution with mean = 0 and standard deviation = 10.
ex1_3 = np.random.normal(0,10, 1000)

#1000 samples of beta distribution for pairs of parameters (alpha,beta) = (1,1), (2,2),(3,9)
ex1_4_1 = np.random.beta(1,1,1000)
ex1_4_2 = np.random.beta(2,2,1000)
ex1_4_3 = np.random.beta(3,9,1000)

#10000 samples of two-dimensional normal distributions each with different means
mi = np.array([0,0])
cov1 = np.array([[1,0],[0,1]])
cov2 = np.array([[2,1],[1,2]])
x1,y1 = np.random.multivariate_normal(mi,cov1,5000).T
x2,y2 = np.random.multivariate_normal(mi,cov2,5000).T
```

#### Ex2. Plot histograms

```python
_, bins, _ = plt.hist(ex1_1, bins="auto",density=True)
plt.title("Uniform distribution on interval [0,1]")
plt.plot(bins, np.ones_like(bins))
plt.grid()
plt.legend(["Calculated","Analytical"])
plt.show()
```

```python
plt.hist(ex1_2, bins="auto", density=True)
plt.grid()
x = np.linspace(-4,4,1000)
plt.plot(x,stats.norm.pdf(x,0,1))
plt.title("Uniform distribution with mean= 0 and standard deviation = 1")
plt.legend(["Calculated","Analytical"])
plt.show()
```

```python
plt.hist(ex1_3, bins="auto",density=True)
plt.grid()
x = np.linspace(-40,40,1000)
plt.plot(x,stats.norm.pdf(x,0,10))
plt.title("Uniform distribution with mean= 0 and standard deviation = 10")
plt.legend(["Calculated","Analytical"])
plt.show()
```

```python
plt.hist(ex1_4_1, bins="auto", density=True)
plt.grid()
x = np.linspace(0,1,1000)
plt.plot(x,stats.beta.pdf(x,1,1))
plt.title("Uniform distribution alpha,beta=1,1")
plt.legend(["Calculated","Analytical"])
plt.show()
```

```python
plt.hist(ex1_4_2, bins="auto", density = True)
plt.grid()
x = np.linspace(0,1,1000)
plt.plot(x,stats.beta.pdf(x,2,2))
plt.title("Uniform distribution alpha,beta=2,2")
plt.legend(["Calculated","Analytical"])
plt.show()
```

```python
plt.hist(ex1_4_3, bins="auto", density = True)
plt.grid()
x = np.linspace(0,1,1000)
plt.plot(x,stats.beta.pdf(x,3,9))
plt.title("Uniform distribution alpha,beta=3,9")
plt.legend(["Calculated","Analytical"])
plt.show()
```

Ploted calculated functions are similar to those ones analytical.

```python
alpha = np.zeros(1000)
beta = np.zeros(1000)
x = np.zeros(1000)
sigma = np.zeros(1000)
mi = np.zeros(1000)
y = np.zeros(1000)

for i in range(1000):
  alpha[i] = np.random.normal(2,0.1)
  beta[i] = np.random.normal(0.2, 0.2)
  x[i] = np.random.normal(0,3)
  sigma[i] = halfnorm.rvs(loc = 0, scale = 1)
  mi[i] = alpha[i] +beta[i]*x[i]
  y[i] = np.random.normal(mi[i],sigma[i])
```

#### Ex.6

```python
plt.scatter(x,y)
for a,b in zip(alpha,beta):
   x_ = np.linspace(-6,6,100)
   y2 = a +b*x_
   plt.plot(x_,y2,"red",alpha = 0.09)

plt.title("y(x)")
plt.show()
```
